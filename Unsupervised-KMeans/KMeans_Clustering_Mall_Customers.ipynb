```python
from sklearn import cluster
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt


data = pd.read_csv('./Mall_Customers.csv')
print(data.shape)
print(data.columns)
Ndata=data.drop(['CustomerID','Genre','Age'],axis=1)
print(Ndata.head())
plt.figure(1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(Ndata['Annual Income (k$)'],Ndata['Spending Score (1-100)'],'ko')
plt.title('raw data')
```

```python
# elbow method
cl = []
dis = []
for i in range(1,11,1):
    cl.append(i)
    kmModel = cluster.KMeans(n_clusters=i,random_state=10)
    kmModel = kmModel.fit(Ndata)
    dis.append(kmModel.inertia_)

plt.figure(1)
plt.plot(cl,dis,'--ko')
plt.xlabel('clusters')
plt.ylabel('sum of squared distances')
plt.grid('on')
```

```python
#Final number of cluster is 3 from elbow method
kmModel = cluster.KMeans(n_clusters=3,random_state=10)
kmModel = kmModel.fit(Ndata)
print(kmModel.cluster_centers_)
print(kmModel.labels_)
label = kmModel.labels_
centroids = kmModel.cluster_centers_

plt.figure(1,figsize=(9,9))
plt.subplot(2,1,1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(Ndata['Annual Income (k$)'],Ndata['Spending Score (1-100)'],'ko')

plt.title('raw data')
plt.subplot(2,1,2)
plt.scatter(Ndata['Annual Income (k$)'],Ndata['Spending Score (1-100)'],c=label)

plt.title('Segmented Data')
for i in range(len(centroids)):
    plt.plot(centroids[i][0],centroids[i][1],'bd')
```

```python
import pandas as pd
from sklearn import cluster
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

data = pd.read_csv('./Mall_Customers.csv')
print(data.head())
print(data.columns)
NData=data.drop(['CustomerID','Genre','Age'],axis=1)
print(NData.head())
X = NData.values
```

```python
# elbow method
cl = []
dis = []
for i in range(1,11,1):
    cl.append(i)
    kmModel = cluster.KMeans(n_clusters=i,random_state=10)
    kmModel = kmModel.fit(X)
    dis.append(kmModel.inertia_)

plt.figure(1)
plt.plot(cl,dis,'--ko')
plt.xlabel('clusters')
plt.ylabel('sum of squared distances')
plt.grid('on')
```

```python
# final number of clusters is 3
# kmeans models for 3 cluster 
kmModel = cluster.KMeans(n_clusters=3,random_state=10)
kmModel = kmModel.fit(X)
print(kmModel.cluster_centers_)
print(kmModel.labels_)
label = kmModel.labels_
centroids = kmModel.cluster_centers_

plt.figure(2,figsize=(9,9))
plt.subplot(2,1,1)
# markers o-cicle,s-square,d-diamond,*-star
plt.plot(X[:,0],X[:,1],'ko')
plt.title('raw data')
plt.subplot(2,1,2)
plt.scatter(X[:,0],X[:,1],c=label)
plt.title('Segmented Data')
for i in range(len(centroids)):
    plt.plot(centroids[i][0],centroids[i][1],'bd')
```
